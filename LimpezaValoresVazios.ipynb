```python
# transformar em valores vazios durante a importação
valores_vazios = ["?" , "-" , "_"]           # -> considera estes casos como valores vazios
df = pd.read_csv("advertising.csv", na_values = valores_vazios)


# usar o replace, substitui o caracter no dicionário por um campo vazio
df = df.replace ({"?" : np.nan, "-" : np.nan , "_" : np.nan})
```

```python
# remover os valores vazios com (dropna)

# axis (1) remove a coluna inteira, ou (0) remove a linha inteira 
# hall (all) exclui linhas/colunas em que todos os campos estão vazios, ou (any) em que basta um campo estar vazio
df = df.dropna(how = "any", axis = 0)
```

```python
# preenche campo vazio com valor fixo (fillna)

media = df["TV"].mean()
df["TV"] = df["TV"].fillna(media)
display(df)
```

```python
# preenchendo com valor anterior ou seguinte / anterior
df["TV"] = df["TV"].fillna(method="ffill") # -> "ffill" = anterior , "bfill" = seguinte
```

```python
# interpolando (faz uma estimativa do valor que deveria ocupar aquele campo)
df["TV"] = df["TV"].interpolate()
display(df)
```
